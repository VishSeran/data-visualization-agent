```python
from urllib.parse import urlparse

def is_url(string):
    try:
        result = urlparse(string)
        return all([result.scheme, result.netloc])
    except Exception:
        return False


test_strings = [
    "https://www.google.com",
    "http://example.com/path",
    "hello world",
    "example.com"
]

for s in test_strings:
    print( is_url(s))
```
